## Conclusion for EDA analysis


```python
1. Loans related to property investments (e.g., buying land or a home) tend to involve higher credit amounts. These segments can be 
prioritized for high-value loans.
2. Applicants categorized as state servants demonstrate higher credit applications, making them a target group for structured loan 
schemes with attractive repayment terms.
3. Loans for hobbies or third-party needs generally involve lower credit amounts but may reflect a lack of financial planning, 
Banks should scrutinize these applicants carefully.
4. Municipal apartments and housing with parents show lower default rates, indicating stability. These applicants can be prioritized for loans.
5. Women employees demonstrate strong repayment reliability, making them ideal candidates for personal or professional loans.
6. High-value loans for non-essential purposes (e.g., leisure activities or unclassified uses) should involve stricter approval 
criteria to minimize default risks.
```
